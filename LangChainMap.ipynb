```python
# Helper functions

def extract_keywords(text: str, max_keywords: int = 3) -> str:
    """
    Extracts keywords from the text using a simple frequency-based method.
    Returns only the top 3 keywords.
    """
    stopwords = {
        'the', 'and', 'a', 'of', 'to', 'in', 'for', 'on', 'with', 'at',
        'by', 'an', 'is', 'it', 'this', 'that', 'from', 'as', 'are', 'be',
        'was', 'were', 'or', 'but'
    }
    text_clean = re.sub(r'[^\w\s]', '', text).lower()
    words = text_clean.split()
    filtered_words = [word for word in words if word not in stopwords]
    word_counts = Counter(filtered_words)
    most_common = word_counts.most_common(max_keywords)
    keywords = [word for word, _ in most_common]
    return " ".join(keywords)

def is_english(text: str, threshold: float = 0.9) -> bool:
    """
    Determines if the text is likely English by verifying the ratio of ASCII alphabetic characters.
    """
    total_alpha = 0
    ascii_count = 0
    for char in text:
        if char.isalpha():
            total_alpha += 1
            if char in string.ascii_letters:
                ascii_count += 1
    if total_alpha == 0:
        return True
    return (ascii_count / total_alpha) >= threshold

def translate_to_english(text: str) -> str:
    """
    Translates text to English using OpenAI's ChatCompletion API.
    """
    try:
        prompt = f"Translate the following text to English:\n\n{text}"
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a translation assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )
        translation = response['choices'][0]['message']['content']
        logger.info("Translation successful using OpenAI ChatCompletion.")
        return translation.strip()
    except Exception as e:
        logger.error(f"Error during translation: {e}", exc_info=True)
        return text

def translate_and_extract(text: str) -> (str, str):
    """
    Translates the text to English and extracts the top 3 keywords.
    Returns a tuple (translation, keywords).
    """
    try:
        prompt = (
            "Translate the following text to English and extract the top 3 keywords from it. "
            "Return the result as a JSON object with keys 'translation' and 'keywords', "
            "where 'keywords' is a comma-separated string.\n\n"
            f"Text: '''{text}'''"
        )
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a translation and keyword extraction assistant."},
                {"role": "user", "content": prompt}
            ],
            temperature=0
        )
        output = response['choices'][0]['message']['content']
        parsed = json.loads(output)
        translation = parsed.get("translation", text)
        keywords = parsed.get("keywords", "")
        logger.info("Translation and keyword extraction successful.")
        return translation.strip(), keywords.strip()
    except Exception as e:
        logger.error(f"Error during combined translation and extraction: {e}", exc_info=True)
        # Fallback to separate translation and keyword extraction
        translation = translate_to_english(text)
        keywords = extract_keywords(translation)
        return translation, keywords

# News fetching function

def fetch_efficient_news(articles_per_request: int = 100, max_requests: int = 100, last_hours: int = 72, max_retries: int = 3) -> list:
    """
    Fetches news articles published within the last `last_hours` hours,
    using multiple MENA-specific queries to capture diverse articles.
    """
    from_date = (datetime.datetime.utcnow() - datetime.timedelta(hours=last_hours)).isoformat()
    print(f"Fetching news articles from the last {last_hours} hours (since {from_date})")
    all_articles = []
    seen_urls = set()
    request_count = 0
    timestamp = datetime.datetime.utcnow().isoformat()
    search_queries = [
        "Middle East", "North Africa", "MENA",  "Algeria","Bahrain",
        "Egypt","Gaza","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon",
        "Libya","Morocco","Oman","Palestine","Qatar","Saudi Arabia",
        "Syria","Tunisia","United Arab Emirates","Yemen","Sudan"
    ]

    print("Testing NewsAPI connection...")
    test_params = {'q': 'Middle East', 'apiKey': NEWS_API_KEY, 'pageSize': 1}
    try:
        test_response = requests.get(NEWS_API_ENDPOINT, params=test_params)
        test_data = test_response.json()
        if test_response.status_code != 200 or test_data.get('status') == 'error':
            error_message = test_data.get('message', 'Unknown error')
            print(f"⚠️ API Test Failed: {error_message}")
            logger.error(f"API Test Failed: {error_message}")
            return []
        else:
            print("✅ API Test Successful! Beginning article collection...\n")
    except Exception as e:
        print(f"⚠️ API Test Failed with exception: {e}")
        logger.error(f"API Test Failed with exception: {e}", exc_info=True)
        return []

    for query in search_queries:
        if request_count >= max_requests:
            print(f"Reached maximum number of requests ({max_requests}). Stopping.")
            break

        request_count += 1
        print(f"\n--- Request {request_count}/{max_requests}: Fetching articles about '{query}' ---")
        params = {
            'q': query,
            'from': from_date,
            'apiKey': NEWS_API_KEY,
            'pageSize': articles_per_request,
            'language': 'en',
            'sortBy': 'relevancy'
        }
        safe_params = params.copy()
        safe_params['apiKey'] = '***HIDDEN***'
        print(f"Request parameters: {safe_params}")
        attempts = 0
        base_wait_time = 10
        success = False

        while attempts <= max_retries and not success:
            try:
                response = requests.get(NEWS_API_ENDPOINT, params=params)
                response_data = response.json()
                if response_data.get('status') == 'error':
                    error_message = response_data.get('message', 'Unknown API error')
                    print(f"API Error: {error_message}")
                    logger.error(f"API Error: {error_message}")
                    attempts += 1
                    time.sleep(base_wait_time)
                    continue
                response.raise_for_status()
                data = response_data
                if 'articles' not in data:
                    print(f"Unexpected response format: {data}")
                    logger.error(f"Unexpected response format: {data}")
                    break
                articles = data.get('articles', [])
                if not articles:
                    logger.info(f"No articles found for query '{query}'.")
                    print(f"No articles found for query '{query}'.")
                    break

                new_articles_count = 0
                for i, article in enumerate(articles):
                    url = article.get('url', '')
                    if url in seen_urls or not url:
                        continue
                    article['fetched_at'] = timestamp
                    article['request_num'] = request_count
                    article['query'] = query
                    article['position'] = i + 1
                    print(f"{request_count}.{new_articles_count+1}. {article.get('title', 'No Title')}")
                    print(f"   Source: {article.get('source', {}).get('name', 'Unknown')}")
                    print(f"   Published: {article.get('publishedAt', 'Unknown date')}")
                    print(f"   URL: {url}")
                    print(f"   Content: {article.get('content', 'No content snippet available')}")
                    print()
                    all_articles.append(article)
                    seen_urls.add(url)
                    new_articles_count += 1

                logger.info(f"Request {request_count}: Added {new_articles_count} new articles out of {len(articles)} received.")
                print(f"Added {new_articles_count} new unique articles. Total collection: {len(all_articles)}")
                success = True

            except requests.exceptions.HTTPError as e:
                print(f"HTTP Error: {e}")
                logger.error(f"HTTP Error: {e}", exc_info=True)
                if response.status_code == 429:
                    wait_time = base_wait_time * (attempts + 1)
                    logger.warning(f"Rate limit reached. Waiting {wait_time} seconds...")
                    print(f"Rate limit reached. Waiting {wait_time} seconds before retrying...")
                    time.sleep(wait_time)
                    attempts += 1
                elif response.status_code in [401, 403]:
                    print("Authentication or quota error. Stopping requests.")
                    return all_articles
                else:
                    attempts += 1
                    wait_time = base_wait_time * attempts
                    print(f"Retrying in {wait_time} seconds...")
                    time.sleep(wait_time)
            except Exception as e:
                logger.error(f"Unexpected error on request {request_count}: {e}", exc_info=True)
                print(f"Unexpected error: {str(e)}")
                attempts += 1
                time.sleep(base_wait_time * attempts)

        if not success:
            logger.error(f"Exceeded maximum retries for request {request_count}.")
            print(f"Exceeded maximum retries for request {request_count}. Moving on...")
        if request_count < max_requests:
            print("Waiting 1 second before next request...")
            time.sleep(1)

    logger.info(f"Total articles fetched: {len(all_articles)} across {request_count} requests.")
    print(f"\nTotal articles fetched: {len(all_articles)} across {request_count} requests.")
    return all_articles
```

```python
# Import BaseTool from LangChain and create our custom tool wrapper.
from langchain.tools import BaseTool

class NewsFetcherTool(BaseTool):
    # Add type annotations for the class attributes.
    name: str = "NewsFetcherTool"
    description: str = (
        "Fetches news articles from NewsAPI for the MENA region published within a given time interval. "
        "Parameters: query (default is 'MENA'), last_hours, articles_per_request, and max_requests."
    )

    def _run(self,
             query: str = "MENA",
             last_hours: int = 72,
             articles_per_request: int = 100,
             max_requests: int = 5) -> str:
        """
        Runs the news-fetching function and returns the fetched articles as a JSON string.
        """
        articles = fetch_efficient_news(
            articles_per_request=articles_per_request,
            max_requests=max_requests,
            last_hours=last_hours
        )
        return json.dumps(articles, ensure_ascii=False, indent=2)

    async def _arun(self,
                    query: str = "MENA",
                    last_hours: int = 72,
                    articles_per_request: int = 100,
                    max_requests: int = 5) -> str:
        raise NotImplementedError("Asynchronous execution not implemented yet.")
```
